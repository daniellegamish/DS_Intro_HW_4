```python
import requests
import urllib   
import json

#read the file
n = ''
try:
    textFile=input('Please enter your citys in a text file (for example dests.txt): ')
    myfile=open(textFile,'r',encoding='utf-8')
    myfile=myfile.read()
    Lines=myfile.splitlines()
except:
    textFile = n
    print('We are sorry to inform you that we could not find this file.')
    textFile=input('Please try again, enter your citys in a text file (for example dests.txt): ')

if textFile != n:
    api_key=input('Please enter your API: ')

service_url = 'https://maps.googleapis.com/maps/api/distancematrix/json?'
mydictionary=dict()
mydictionary['origins']='Jerusalem'
mydictionary['key'] = api_key
mylist=list()
information=tuple()
i=0
all_dis=list()
all_cities_names=list()
```

```python
for city in Lines:
    #Request and receive data from web pages
    #Integrity check 
    mydictionary['destinations'] = city
    try:
        url1 = service_url + urllib.parse.urlencode(mydictionary)
        response=requests.get(url1)
        address=city
        url2="https://maps.googleapis.com/maps/api/geocode/json?address=%s&key=%s" % (address, api_key)
        response2 = requests.get(url2)
        if response.status_code != 200 or response2.status_code !=200 :
            print("HTTP Error")
        else:
            response=response.json()
            response2=response2.json()
            
            #saving the data in variables
            km=response['rows'][0]['elements'][0]['distance']['text']
            time=response['rows'][0]['elements'][0]['duration']['text']
            latitude=response2['results'][0]['geometry']['location']['lat']
            longitude=response2['results'][0]['geometry']['location']['lng'] 
            
            
            if time.find('day') > 0:
                index_Day = time.find('day')-1
                H = (int(time[:index_Day]))*24
                if time.find('hours') > 0:
                    index_Hours = time.find('hours')-1
                    h_time = int(time[index_Day+5 : index_Hours])
                    h_time += H
                    time = str(h_time) + " hours"   
                else:    
                    time = str(H) + "hours"
                    
            #make a list with all the cities
            dictionary=dict()
            mylist.append(dictionary)
            information=('distance:'+ km, 'time:' + time, 'longitude:', longitude, 'latitude:' ,latitude)
            dictionary[city]=information
            print("The destination is: " + city)
            print(" the distance is: " + km)
            print( " Duration of arrival to the destination: " + time)
            print( " The longitude are: "+ str(longitude) )
            print(" The latitude are: "+ str(latitude))
            print()
            
            #make a list with all the distances
            #make a list with all the cities names
            all_dis.append(km)
            all_cities_names.append(city)
            i+=1
    except:
        print("Requests error")
```

```python
#Choose the three cities with the largest distance from Jerusalem and print them
if len(mylist)>0:
    the_3_Cities_for_dict = ''
    the_3_Cities=list()   
    for km_num in all_dis:
       while len(the_3_Cities)<3 : 
           big_num=max(all_dis) 
           index=all_dis.index(max(all_dis))
           the_3_Cities.append(all_cities_names[index])
           the_3_Cities_for_dict += all_cities_names[index] + " "
           all_dis[index]= '0'        
    dictionary['The three cities'] = the_3_Cities_for_dict  
    print("The three cities with the largest distance from Jerusalem are: " + dictionary['The three cities']) 
```
